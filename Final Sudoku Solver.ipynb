```python
def solveblock(blockind):

    tlc = thisdict[blockind]
    inbox = [] # lists all the numbers in the box 
    noind = [] #lists array indices where there is 0
    for i in range(3):
        for j in range(3): 
            #print(inp1[tlc[0]+i][tlc[1]+j]) # lists all the numbers in the box 
            
            if inp1[tlc[0]+i][tlc[1]+j] == 0:
                #print(tlc[0]+i, tlc[1]+j) #lists array indices where there is 0
                noind.append([tlc[0]+i, tlc[1]+j])
            else:
                inbox.append(inp1[tlc[0]+i][tlc[1]+j]) #lists numbers that are present in the box
    
    
    r = [i for i in range(1,10)]
    
    notinbox = [x for x in r if x not in inbox] # -- numbers that are not present in the box
    notinbox = np.array(notinbox)
    
    
    poss = []  # list of possibile numbers at each empty cell in the block
    #### the possibilities for a cell is determined by checking the row and column to which the cell belongs ###
    for i in noind:
        posst = []
        
        for j in notinbox:
            
            if j not in inp1[i[0]] and j not in inp1[:,i[1]]:

                posst.append(j)
        poss.append(posst)
        

    # noind -- indices where empty/0 places; first row (left to right), then column
    # notinbox -- numbers that are not present in the box
    # poss -- possibilities for each index/cell
    # noind --- poss ; ie. they map to each other
    
    rem = [] # solutions we are going to find in the next step
    remind = [] #indices/cells whose solutions we are going to find in the next step
    
    ### METHOD 1 to find solution: if there is only one value a cell can accept, then that value is the solution for the cell
    
    for i in range(len(poss)):
        
        if len(poss[i]) == 1:
            
            inp1[noind[i][0]][noind[i][1]] = poss[i][0]  # replaces the empty cell with the number
            rem.append(poss[i][0]) 
            remind.append(noind[i])
   
    notinbox1 = [x for x in notinbox if x not in rem] # --- updated the numbers that are not present in the box
       
    noind1 = [x for x in noind if x not in remind] # ---- updated the cells that are empty
       

    ##### check possibilities for the updated cells ####
    poss1 = []  # list of possibilities for each of the empty boxes
    for i in noind1:
        posst = []
        
        for j in notinbox1:
            
            if j not in inp1[i[0]] and j not in inp1[:,i[1]]:
                
                posst.append(j)
        poss1.append(posst)
        

    ### noind1 -- updated empty cells
    ### notinbox1 -- updated numbers that are not present in the box
    ### poss1 -- updated possibilities for each cell
    

    ### check using the only possibility method
    ### first count the number of times a possible number for empty cells occurs in the list of possibilities for each cell

    
    ### METHOD 2 : checks for numbers that can only occur in one cell in the block
    
    freq1 = [] ### possible numbers that occur only once
    for j in notinbox1:
        k = sum(x.count(j) for x in poss1)
        if k == 1:
            freq1.append(j)
    

    freq1ind = [] ### indices where those only-once occuring numbers are
    for i in freq1:
        for j in range(len(poss1)):
            if i in poss1[j]:
                freq1ind.append(noind1[j])

    #print(freq1ind) ### 
    ### freq1 maps with freqind1            

    for i in range(len(freq1)):
        inp1[freq1ind[i][0]][freq1ind[i][1]] = freq1[i]  ### replace the cells

    #print(inp1)
    ## This finishes the iteration for this block.
```
